```python
import pandas as pd
data1 = pd.read_csv('/home/data/MSA8010/msa8010f16t03/data/train_numeric.csv')
```

```python
# Counting number of rows in the data frame:
c=data1.Response.count()
```

```python
#Counting of filled values in each column in dataframe:
count1=[]
count1=data1.count()
df = pd.DataFrame(count1)
df.columns=['filled_count']
df['total_count']=c
df2=df.reset_index()
df3=df2.rename(columns={'index':'Columns'})
```

```python
# Replacing all non null values with response variables
data3=data1
for column in data1:
    data3.loc[data3[column].notnull(), column] = data3['Response']
       
count2=[]
count2=data3.sum()
dfn= pd.DataFrame(count2)
dfn.columns=['Success']
dfn2=dfn.reset_index()
dfn3=dfn2.rename(columns={'index':'Columns'})
```

```python
#merging tables
x=pd.merge(df3, dfn3, on='Columns', how='inner')
```

```python
#Creating new variables
x['filled_perc']=x['filled_count']/x['total_count']*100
x['success_perc_filled']=x['Success']/x['filled_count']*100
x2 = x.ix[1:]
```

```python
#exporting to csv
x2.to_csv('/home/ajain13/kaggle/summary.csv', sep=',', encoding='utf-8')
```
